```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
### Modify This  
Filenames = ['hpcc_amd20.csv','hpcc_Intel18.csv','PC_AMD7.csv']
CPU = ['amd20','intel18','amd7']
peaks = [2.6,2.4,2.3] # Theoretical peak performance 
###
```

```python
n=0
peak=np.array([[10, 10000], [peaks[n], peaks[n]]])

df = pd.read_csv(Filenames[n])

plt.figure(figsize=(10, 5))

plt.plot(df['N'], df['gflops'],'b', marker='o',label=CPU[n])
plt.plot(peak[0],peak[1],'r--',label='Theoretical peak performance')


plt.title('Gflop/s vs. N')
plt.xlabel('N')
plt.ylabel('GFLOP/s')
plt.xscale('log')
plt.yscale('log')
plt.grid(True)
plt.legend()
plt.show()
```
