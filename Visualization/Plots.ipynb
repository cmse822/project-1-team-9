```python
import pandas as pd
import matplotlib.pyplot as plt

Filenames = ['PC_AMD7_Default.csv', 'PC_AMD7_O3.csv', 'hpcc_Intel18_Default.csv', 'hpcc_Intel18_O3.csv']

dfs = []

for filename in Filenames:
    df = pd.read_csv(filename)
    df['Arithmetic_Intensity'] = df['average_elapsed_time'] * df['gflops'] / (3 * 8)
    dfs.append(df)

plt.figure(figsize=(10, 5))

for i, df in enumerate(dfs):
    plt.plot(df['Arithmetic_Intensity'], df['gflops'], marker='o')

plt.title('Performance vs Arithmetic_Intensity')
plt.xlabel('GFLOP/Bytes')
plt.ylabel('GFLOP/s')
plt.xscale('log')
plt.yscale('log')
plt.grid(True)
plt.legend(Filenames)
plt.show()
```
